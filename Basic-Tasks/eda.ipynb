### Importing the Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Loading the Data

```python
# Reading the CSV file
df = pd.read_csv("Iris.csv")
df.head()
```

### Basic dataset info

**Checking the shape**

```python
df.shape
```

**Observation**
<li>We can see that the dataframe contains 6 columns and 150 rows.</li>


**Columns Info**

```python
df.info()
```

**Observation**
<li>Here, We can see that only one column has categorical data and all the other columns are of the numeric type with non-Null entries.</li>


**Dataset statistics**

```python
df.describe()
```

**Observation**
<li>We can see the count of each column along with their mean value, standard deviation, minimum and maximum values.</li>


**Check for null values**

```python
df.isnull().sum()
```

**Observation**
<li>We can see that no column as any missing value.</li>

### Distribution Plot (Histogram) - Sepal Length

```python
sns.histplot(df['SepalLengthCm'], kde=True)
plt.title('Distribution of Sepal Length')
plt.show()
```

**Observation**
<br>
<li>Sepal length has a roughly normal distribution with a peak around 5.5 cm.</li>
<li>There are a few samples with higher sepal lengths, but no extreme outliers.</li>
<li>The distribution is slightly skewed to the right.</li>
<li>It follows a Platykurtic distribution






### Distribution Plot (Histogram) - Sepal Width

```python
sns.histplot(df['SepalWidthCm'], kde=True)
plt.title('Distribution of Sepal Width')
plt.show()
```

**Observation**
<br>
<li>It follows a normal distribution</li>
<li>Sepal width shows a slight right skew with most values between 2.5 and 3.5 cm.</li>
<li>There are a few samples with small and large widths, but no extreme outliers.</li>
<li>The mode is around 3.0 cm.</li>



### Visualizing the target column

```python
sns.countplot(x='Species', data=df, hue ='Species', palette = 'hls')
plt.show()
```

**Observation**
<br>
<li>The dataset is balanced, with 50 samples for each species.</li>
<li>No species is overrepresented or underrepresented.</li>
<li>This balanced distribution is ideal for classification tasks.</li>




### Box Plot - Sepal Length Across Species

```python
sns.boxplot(x='Species', y='SepalLengthCm', data=df)
plt.title('Box Plot of Sepal Length Across Species')
plt.show()
```

**Observation**
<br>
<li>Iris-setosa has the smallest sepal length among the three species.</li>
<li>Iris-virginica has the largest sepal length range, with some outliers.</li>
<li>The median sepal length for Iris-virginica is significantly higher than the other species.</li>


### Box Plot - Sepal Width Across Species

```python
sns.boxplot(x='Species', y='SepalWidthCm', data=df)
plt.title('Box Plot of Sepal Width Across Species')
plt.show()
```

**Observation**
<br>
<li>Iris-setosa tends to have the widest sepals compared to other species.</li>
<li>Iris-versicolor and Iris-virginica have similar sepal widths.</li>
<li>There are a few outliers in Iris-virginica for sepal width.</li>



### Scatter Plot - Sepal Length vs Sepal Width

```python
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm', hue='Species', data=df)
plt.title('Sepal Length vs Sepal Width')
plt.show()
```

**Observation**
<br>
<li>Iris-setosa clusters at the lower range sepal length.</li>
<li>Iris-versicolor and Iris-virginica overlap, making them harder to distinguish based on sepal features alone.</li>
<li>Sepal dimensions provide decent but not perfect separation between species.</li>


### Scatter Plot - Petal Length vs Petal Width

```python
sns.scatterplot(x='PetalLengthCm', y='PetalWidthCm', hue='Species', data=df)
plt.title('Petal Length vs Petal Width')
plt.show()
```

**Observation**
<br>
<li>Iris-setosa is clearly distinguishable from the other species, with much smaller petal dimensions.</li>
<li>Iris-versicolor and Iris-virginica show some overlap but are better separated compared to sepal dimensions.</li>
<li>The relationship between petal length and petal width is linear for all species.</li>



### Violin Plot - Petal Length Across Species

```python
sns.violinplot(x='Species', y='PetalLengthCm', data=df)
plt.title('Violin Plot of Petal Length Across Species')
plt.show()
```

**Observation**
<br>
<li>Iris-setosa shows the least variation in petal length.</li>
<li>Iris-virginica has the largest variation, with a wide range of petal lengths.</li>
<li>Petal length helps separate species, especially distinguishing Iris-setosa from the other two.</li>


### Correlation Heatmap

```python
corr = df.drop(columns=['Id','Species']).corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Sepal and Petal Attributes')
plt.show()
```

**Observation**
<br>
<li>PetalLengthCm and PetalWidthCm have a very high correlation (0.96).</li>
<li>SepalLengthCm has a moderate correlation with PetalLengthCm and PetalWidthCm.</li>
<li>SepalWidthCm shows very little correlation with other features.</li>


### Pairplot

```python
sns.pairplot(df.drop(columns=['Id']), hue="Species", palette='flare')
plt.suptitle("Pairplot of Sepal and Petal Attributes", y=1.02)
plt.show()
```

**Observation**
<br>
<li>Petal features show clearer separation between species compared to sepal features.</li>
<li>Iris-setosa is very distinct and easily separable.</li>
<li>There is considerable overlap between Iris-versicolor and Iris-virginica in most plots, except for petal dimensions.</li>


### Swarm Plot - Petal Width Across Species

```python
sns.swarmplot(x='Species', y='PetalWidthCm', data=df, hue='Species', palette='hls')
plt.title('Swarm Plot of Petal Width Across Species')
plt.show()
```

**Observation**
<br>
<li>Iris-setosa has significantly smaller petal width compared to the other two species.</li>
<li>Iris-virginica generally has the largest petal width, with very few overlaps with Iris-versicolor.</li>
<li>There is a slight overlap between Iris-versicolor and Iris-virginica, but the petal width is still a good feature for distinguishing species.</li>
